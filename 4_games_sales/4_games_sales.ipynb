Попробуем аналогичным способом обработать пропуски в столбцах `critic_score`, `user_score` и `rating`. Такое решение хоть и не идеальное, но допустимое поскольку:

* Как правило возрастной рейтинг для одной и той же игры не зависит от конкретной платформы, при условии, что игровой контент идентичен (на него влияют скорее законодательные ограничения той или иной страны)


* Оценки пользователей и критиков одной и той же игры могут кардинально отличаться на разных платформах только, если на одной из них присутствуют серьезные баги, или игра под одним тайтлом выходит на консолях очень разных поколений (с разными техническими характеристиками), из-за чего внутриигровой контент может серьезно различаться. Но такие случаи довольно редки.  

```python
# Функция замены пропуска рейтингом данной игры на других платформах 
# Рейтинг - переменная категориальная, поэтому функция возвращает рейтинг первой игры из Series 
def fill_rating(row):
    rating = row['rating']
    name = row['name']
    if str(rating) == 'nan':
        data = df[df['name'] == name].dropna(subset=['rating'])['rating'].reset_index(drop=True)
        if len(data) != 0:
            our_rating = data.loc[0]
            return our_rating
        else:
            return rating
    else:
        return rating
```

```python
# Функция замены пропуска оценкой пользователей данной игры на других платформах 
# Функция возвращает среднее значение оценки игры на других платформах
def fill_user_score(row):
    score = row['user_score']
    name = row['name']
    if str(score) == 'nan':
        data = df.query('name == @name and user_score != "tbd"')['user_score'].reset_index(drop=True)
        if len(data) != 0:
            data = data.astype('float')
            our_score = data.mean()
            return our_score
        else:
            return score
    else:
        return score
```

```python
# Функция замены пропуска оценкой критиков данной игры на других платформах 
# Функция возвращает среднее значение оценки игры на других платформах
def fill_critic_score(row):
    score = row['critic_score']
    name = row['name']
    if str(score) == 'nan':
        data = df[(df['name'] == name) & (df['user_score'] != 'tbd')].dropna(subset=['critic_score'])['critic_score'].reset_index(drop=True)
        if len(data) != 0:
            our_score = data.mean()
            return our_score
        else:
            return score
    else:
        return score
```

```python
df['rating'] = df.apply(fill_rating, axis=1)
```

```python
df['user_score'] = df.apply(fill_user_score, axis=1)
```

```python
df['critic_score'] = df.apply(fill_critic_score, axis=1)
```

```python
# Проверка результата
df[['critic_score', 'user_score', 'rating']].info()
```

Оставшиеся ропуски в столбцах `critic_score`, `user_score` и `rating`  невозможно объективно заполнить каким-либо значением. Поэтому пока оставим их без изменений и не будем удалять из таблицы - поскольку в других столбцах содержится информация, которая будет необходима для дальнейшего исследования. 

Отдельно отметим, что в столбце `user_score` помимо пропусков присутствует строковое значение "tbd". Его обработкой займемся на следующем шаге.

По этой причине, замена всех пропусков и "tbd" нулями в соответвтвующих столбцах, не должна сильно повлиять на дальнейший ход исследования - мы просто не будем включать игры с нулевой оценкой в итоговые  результаты исследования.

Также проверим нормально ли распределены наши выборки. Для проверки используем тест Шапиро-Уилка, зададим уровень значимости $\alpha = 0.05
  $. 

```python
# Проверка на нормальность распределения XOne
st.shapiro(user_df.query('platform == "XOne"')['user_score'])
```

```python
# Проверка на нормальность распределения PC
st.shapiro(user_df.query('platform == "PC"')['user_score'])
```

Значение pvalue в обоих случаях сильно ниже установленного уровня значимости. Таким образом, результаты теста Шапиро-Уилка свидетельствуют о том, что первоначальное предположение судя по всему верно - обе выборки вероятнее всего не являются нормально распределенными. Однако,  благодаря ЦПТ, если размер выборки достаточно большой, выборочные средние, которые можно получить из одной и той же генеральной совокупности, должны быть распределены нормально вокруг истинного среднего этой совокупности.

По этой причине, так как выборки независимы, проверка будет осуществляться при помощи метода `st.ttest_ind`. Учитывая, что размер выборок сопоставим друг с другом и выборочные дисперсии различаются не сильно, установим параметр `equal_var=True`

Итак, в нашем случае нулевая гипотеза будет о равенстве средних оценок пользователей, а  альтернативная о том, что  они отличаются. 

$\begin{equation*}
 \begin{cases}
   H_0 :\text{Средние пользовательские рейтинги платформ Xbox One и PC равны}\\
   H_1 :\text{Средние пользовательские рейтинги платформ Xbox One и PC одинаковые различаются }
 \end{cases}
\end{equation*}$

Установим уровень значимости  $\alpha = 0.05$



```python
# Проверка двусторонней гипотезы о равенстве среднего независимых выборок
alpha = 0.05
results = st.ttest_ind(user_df.query('platform == "PC"')['user_score'], user_df.query('platform == "XOne"')['user_score'])

print('p-значение: ', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Так как распределение оценок пользователей все-таки отличается от нормального, а размер выборок не так уж велик, на всякий случай проверим исходную гипотезу припомощи непараметрического теста Манна-Уитни, для проведения которого отсутсвует требование о нормальности распределения. 

```python
# Проверка двусторонней гипотезы о равенстве среднего при помощи непараметрического теста Манна-Уитни
alpha = 0.05
results = st.mannwhitneyu(user_df.query('platform == "PC"')['user_score'], user_df.query('platform == "XOne"')['user_score'])

print('p-значение: ', results.pvalue)
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Таким образом, по результатам обоих тестов нет оснований отвергать гипотезу о том, что средние пользовательские рейтинги платформ Xbox One и PC не отличаются.

По этой причине проверим гипотезу при помощи двух тестов - параметрического и непараметрического. 

 Первая проверка будет осуществляться при помощи метода `st.ttest_ind`. Учитывая, что размер выборок довольно сильно различается, установим параметр `equal_var=False`

Нулевая гипотеза будет о равенстве средних оценок пользователей, а  альтернативная о том, что  они отличаются.

 \begin{cases}

 \end{cases}
\end{equation*}$

Установим уровень значимости  $\alpha = 0.05$

Таким образом у нас нет оснований отвергать альтернативную гипотезу о том, что средние пользовательские рейтинги жанров Action и Sports отличаются
