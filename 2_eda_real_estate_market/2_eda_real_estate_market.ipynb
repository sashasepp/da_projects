## 2.1 Заполнение пропусков <a class="anchor" id="chapter2.1"></a>

Пропущенные значения были обнаружены в следующих столбцах: `ceiling_height, floors_total, living_area, is_apartment, kitchen_area, balcony, locality_name, airports_nearest, cityCenters_nearest, parks_around3000, parks_nearest, ponds_around3000, ponds_nearest, days_exposition`.


Доля пропусков в столбцах  `floors_total` и `locality_name` крайне незначительна:

```python
print(len(df[df['floors_total'].isna() == True]) / len(df))
len(df[df['locality_name'].isna() == True]) / len(df)
```

В обоих случаях она составляет меньше 1%. Адекватная замена пропущенных значений здесь невозможна. По этой причине для избавления от пропусков целесообразным выглядит удаление соответствующих строк - их отсутствие не должно сильно исказить результаты исследования. 

```python
# удаление строк с пропусками
df.dropna(subset=['floors_total', 'locality_name'], inplace=True)
```

```python
# проверка результата
df.info()
```

Пропусков в столбцах `ceiling_height`, `living_area`, `kitchen_area`,  `is_apartment` и `balcony` уже значительно больше, поэтому их невозможно удалить без ущерба для качества исследования. Все эти столбцы заполняются вручную самими пользователями.

`is_apartment` - столбец с булевым типом данных, он содержит информацию является ли жилье апартаментами. Можно предположить, что если квартира действительно имеет особый статус, то пользователь отдельно отметит это в объявлении. По этой причине пропуски в этом столбце можно связать с тем, что жилье апартаментами не является, т.к. пользователь явно не указал это в своем объявлении. Тогда оптимальный вариант - это замена всех пропусков на `False`:

```python
# Замена всех пропусков столбца is_apartment на False
df['is_apartment'].fillna(False, inplace=True)
```

```python
# Проверка результата
df['is_apartment'].value_counts()
```

Схожая ситуация и со столбцом `balcony` (количество балконов). Раз пользователь явно не указал их количество, то можно предположить, что балконов в его квартире нет вовсе. Раз так, то заполним все пропущенные значения нулями:  

```python
# Заполняем нулями пропуски в balcony
df['balcony'].fillna(0, inplace=True)
```

```python
# Проверка результата
df['balcony'].value_counts()
```

Пропуски в `ceiling_height` можно заполнить медианным значением высоты потолков из данного столбца. Однако высота потолков сильно зависит от характеристик конкретной недвижимости. Одним из факторов, влияющих на этот параметр является этажность здания. По этой причине, чтобы чрезмерно не усреднять данные и меньше повлиять на итоговое распределение, заполним пропуски медианными значениями, расчитанными в зависимости от этажности конкретного дома.


Для начала проверим, насколько корректно расчитываются медианы для домов в разной этажностью:

```python
# создаем сводную таблицу медианных значений высоты потолков в зависимости от этажности
floors_pivot = df.pivot_table(index='floors_total', values='ceiling_height', aggfunc='median')
floors_pivot
```

Привлекает внимание странное значение для 36-этажных домов. Взглянем на эти чудо-квартиры подробнее:

```python
df.query('floors_total == 36')
```

В срезе всего 2 дома без пропущенных значений, поэтому медиана - это среднее арифметическое от обычного и явно завышенного значения. Судя по всему здесь произошла опечатка, и точку поставили не в том месте.

Поэтому заменим "битое" значение на 2.7:

```python
df.loc[5807, 'ceiling_height'] = 2.7
```

```python
# Заполняем пропуски медианным значеним высоты потолков, в зависимости от этажности здания
df['ceiling_height'].fillna(df.groupby(by='floors_total')['ceiling_height'].transform('median'), inplace=True)
```

```python
# Проверка успешности операции
df['ceiling_height'].isna().sum()
```

Остался один пропуск, видимо данных для рассчета медианы не хватило, оставшееся значение заполним медианой высоты потолков, расчитанной по всем квартирам из таблицы

```python
df['ceiling_height'].fillna(df['ceiling_height'].median(), inplace=True)
```

```python
# Заключительная (надеюсь) проверка
df['ceiling_height'].isna().sum()
```

Жилая площадь квартиры в столбце `living_area` также зависит от других переменных, в частности от общей площади жилья. Поэтому пропуски в этом столбце можно заполнить средним соотношением жилой площади `living_area` к общей площади квартиры `total_area`:

```python
# Переменная соотношения жилой площади к общей
area_ratio =  df['living_area'].mean() / df['total_area'].mean()
area_ratio
```

```python
# Заполнение пропусков полученным соотношением
df['living_area'].fillna(df['total_area'] * area_ratio, inplace=True)
```

```python
# Проверка результата
df['living_area'].isna().sum()
```

Аналогичную операцию проделаем и со столбцом `kitchen_area`. Однако здесь площадь кухни зависит не только от общей площади квартиры, но и от того, сколько в ней комнат - например, в студии кухни может не быть вовсе. 

Проверим есть ли у студий заполненные значения для `kitchen_area`

```python
# Количество студий
len(df[df['studio'] == True])
```

```python
# Проверка пропущенных значений  в kitchen_area для квартир-студий
df[df['studio'] == True]['kitchen_area'].isna().sum()
```

Действительно, во всех студиях не указана площадь кухни. 
По этой причине, заполним пропуски средним соотношением площади кухни `kitchen_area` к общей площади `total_area`, для всех квартир кроме тех, которые являются студиями. Для них заменим пропуски нулями

```python
# Переменная соотношения площади кухни к общей площади квартиры
kitchen_ratio = df['kitchen_area'].mean() / df['total_area'].mean()
kitchen_ratio
```

```python
# Заполнение пропусков для всех квартир полученным соотношением
df['kitchen_area'].fillna(df['total_area'] * kitchen_ratio, inplace=True)
```

```python
# Присваивание студиям площади кухни равной нулю
df['kitchen_area'].where(df['studio'] == False, 0, inplace=True)
```

```python
# Проверка на пустые значения
df['kitchen_area'].isna().sum()
```

```python
# Проверка значений kitchen_area отдельно для студий
df.query('studio == True').head()
```

Столбец `days_exposition` содержит информацию о том, сколько дней было размещено объявление (от публикации до снятия). Пропуски здесь могут быть из-за того, что объявление действующее и еще не было снято с показа. Поэтому пропуски в этом столбце оставим без изменений.


Данные в столбцах `airports_nearest`, `cityCenters_nearest`, `parks_around3000`, `parks_nearest`, `ponds_around3000`, `ponds_nearest`, получены автоматически на основе картографических данных.  

Проверим существует ли взаимосвязь между пропусками в этих столбцах. Для начала посмотрим на столбец `airports_nearest`:

```python
df.query('airports_nearest.isna() == True', engine='python').head(10)
```

Как мы видим присутствуют одновременные пропуски во всех вышеобозначанных столбцах. Можно обнаружить закономерность - почти во всех случаях пропуски наблюдаются вне Санкт-Петербурга. Возможно данные в столбцах автоматически расчитываются только для этого города, а в остальных случаях возникают пропуски.

Проверим, правда ли это так: 

```python
# Создаем список data - с названиями нужных столбцов
# Проверяем долю пропусков вне Санкт-Петербурга
data = ['airports_nearest', 'cityCenters_nearest', 'parks_around3000', 'parks_nearest', 'ponds_around3000', 'ponds_nearest']
for column in data:
    print(f'{column}:',  
          len(df[(df[column].isna() == True) &
                 (df['locality_name'] != 'Санкт-Петербург')]), 
          'из', len(df[df[column].isna() == True]))
```

Почти все пропущенные значения действительно встречаются за пределами Санкт-Петербурга. Значит пропуски действительно могут быть связаны с отсутствием информации об этих объектах за пределами Петербурга.

Тем не менее, пропуски присутствуют и в данных, связанных с квартирами в Питере:

[К содержанию](#begining)

Для проверки гипотезы о зависимости цены от того, на каком этаже расположена квартира, необходимо добавить еще один столбец, в котором каждой квартире будет присвоена соответствующая категория: **первый**, **последний** или **другой**

```python
# Функция присваивающая квартире категорию, в зависимости от этажа
def floor_category(row):
    if row['floor'] == 1:
        return 'first'
    elif row['floor'] == row['floors_total']:
        return 'last'
    return 'other'
```

1. Согласно полученным данным, самая высокая стоимость жилья наблюдается в Санкт-Петербурге - в среднем чуть больше 104 тыс. рублей за квадратный метр, а самая низкая в Выборге - около 57 тыс. рублей.


2. Было установлено, что **на стоимость квадратного метра в наибольшей степени влияют следующие факторы**:


* **Общая площадь квартиры**: здесь наблюдается обратная зависимость: при возрастании метража квартиры, стоимость квадратного метра уменьшается.


* **Удаленность от центра**:  зависимость цены от расстояния центра существует, однако выражена она слабее, чем, к примеру, от площади квартиры 


* **Этаж**: было обнаружено, что стоимость квартир на первом и последнем этаже несколько ниже, чем на других. Особенно сильна разница для первых этажей - цена за квадратный метр таких квартир почти на 17 тыс. меньше, чем на любом другом непоследнем этаже.


* **Месяц**:  в течение  года присутствует некоторая цикличность ценообразования - стоимость кавартир постепенно снижается к началу лета, резко дорожает к концу лета и вновь снижается ближе к концу календарного года.


* **Год**: с каждым годом  цены на жилье неизменно растут по причине инфляции. 


3. По результатам **сравнения факторов влияющих на ценообразование в центре Санкт-Петербурга и в других районах** можно сделать следующие выводы:

*  Для всей таблицы наблюдается обратная зависимость: при возрастании метража квартиры, чтоимость квадратного метра уменьшается. Для центра Петербурга обратная зависимость этих параметров сохраняется, однако выражена она уже гораздо слабее.


* В центре Петербурга цена кв. метра несколько больше зависит от количества комнат, нежели за его пределами. Однако в обоих случаях эту связь нельзя назвать значительной. Возможно это связано с тем, что в историческом центре Питера довольно много бывших коммуналок, переделанных со временем в квартиры (как правило в 1-комнатные или студии). По этой причине корреляцонная связь здесь обратная - чем меньше комнат, тем дороже стоит 1 кв. метр жилья.


*  В центральной зоне Петербурга зависимость цены квадратного метра от удаленности от центра практически полностью отсутствует. Видимо ввиду того, что все квартиры из второй выборки итак уже находятся в центре,  на ценообразование в данном районе влияют уже совсем иные факторы.


* В целом для центра Питера сохраняется закономерность, применимая ко всей таблице - квартиры на первом и последнем этажах стоят дешевле, чем на других. Однако для центральной зоны разброс цен чуть менее существенный (во всяком случае в процентном соотношении), нежели за ее пределами.


* Центру Петербурга также как и другим районам свойственна некоторая цикличность ценообразования, однако в центральной зоне более отчетливо выделяются резкие скачки роста и падения цен - в марте, августе (резкий рост) и апреле, сентябре (резкое падение) соответственно.


* Инфляция оказывает схожее влияние на ценообразование как в центре, так и в других районах - везде жилье дорожает от года к году. 


* Как в центре Питера, так и в других населенных пунктах цена квартиры оказывается выше, если в ней существует возможность свободной планировки.


* Сильной взаимосвязи между ценой квартиры и размером кухни или жилой площади не наблюдается ( при этом в центре Питера связь цены и размера кухни все же прослеживается чуть более отчетливо).
